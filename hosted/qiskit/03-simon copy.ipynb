```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import numpy as np
from scipy.linalg import null_space

def simon_oracle(n, s):
    qc = QuantumCircuit(2 * n)
    for i, bit in enumerate(reversed(s)):
        if bit == '1':
            qc.cx(i, n + i)    
    return qc.to_gate(label="Simon Oracle")

def simon_algorithm(n, s):
    """ Implementación del algoritmo de Simon en Qiskit """
    qc = QuantumCircuit(2 * n, n)
    qc.h(range(n))
    oracle = simon_oracle(n, s)
    qc.append(oracle, range(2 * n))
    qc.h(range(n))
    qc.measure(range(n), range(n))
    return qc

def solve_simon(n, s, shots=1024):
    qc = simon_algorithm(n, s)
    backend = Aer.get_backend('aer_simulator')
    transpiled_qc = transpile(qc, backend)
    result = backend.run(transpiled_qc, shots=shots).result()
    counts = result.get_counts()
    equations = np.array([list(map(int, key)) for key in counts.keys()])
    s_solution = null_space(equations % 2)%2
    # Recorrer las soluciones, cada una es una columna en 'solutions'
    summed_solution = np.zeros(s_solution.shape[0], dtype=int)  # Inicializar un vector de ceros
    
    for i in range(s_solution.shape[1]):  # Cada columna es una solución
        summed_solution = (summed_solution + s_solution[:, i].astype(int)) % 2  # Sumar en módulo 2

    summed_solution = summed_solution ^ 1
   
    s_str = ''.join(map(str, summed_solution))
    return counts, s_str, qc


n = 3
s = "101" 
counts, found_s, qc = solve_simon(n, s)
print("\nResultados de las mediciones:")
print(counts)
print(f"\nEl valor secreto s encontrado es: {found_s}")

plot_histogram(counts)
```
